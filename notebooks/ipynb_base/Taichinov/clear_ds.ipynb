```python
import pandas as pd
import string
import nltk
from nltk.tokenize import word_tokenize
```

```python
rag_dataset = pd.read_csv('full_dataset.csv')
rag_dataset
```

```python
rag_dataset.isnull()
```

Удаление строк с пустыми значениями

```python
cleared_ds = rag_dataset.dropna()
```

```python
cleared_ds
```

Удаление дубликатов

```python
cleared_ds = cleared_ds.drop_duplicates(keep=False, ignore_index=True, subset=['context'])
cleared_ds = cleared_ds.drop_duplicates(keep=False, ignore_index=True, subset=['question'])
cleared_ds = cleared_ds.drop_duplicates(keep=False, ignore_index=True, subset=['answer'])
```

```python
cleared_ds
```

```python
cleared_ds = cleared_ds.dropna()
```

```python
cleared_ds['context'] = cleared_ds['context'].apply(lambda x: x.replace('\n', ' '))
```

```python

stop_words = set(nltk.corpus.stopwords.words('english') + ['-', '-', '–','&'])
punctuation = set(string.punctuation)
```

```python
cleared_ds['context'] = cleared_ds['context'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words) and word not in punctuation]))
```

Сохранение датасета

```python
cleared_ds.to_csv('../tf_idf/cleared_dataset.csv')
```
